```python
#   5   .
#        
#         
#       
#       

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('SalahYunus_Clean.csv')


# 1
df.iloc[:, 10:] = df.iloc[:, 10:].replace({'tick': True, 0: False})

df_september = df[df['month'] == 9]

numeric_columns = df_september.iloc[:, 10:].apply(pd.to_numeric, errors='coerce')

resp = numeric_columns.sum().nlargest(5)

resp.plot(kind='bar', color='crimson', edgecolor='black')
plt.title('Top 5 Courses with Most Responses in September')
plt.xlabel('Courses')
plt.ylabel('No. of responses')
plt.xticks(rotation=50, ha='right')
plt.show()

# 2
country_responses = df['Who told you about Genius?'].value_counts().nlargest(10)

plt.figure(figsize=(8, 8))
plt.pie(country_responses, labels=None, startangle=90, autopct='%0.0f%%')
plt.legend(country_responses.index, title='Source', loc='upper right')
plt.title('Most popular sources leading to Genius')
plt.show()


# 3
country_responses = df['Country'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(country_responses, labels=None, startangle=90)
plt.legend(country_responses.index, title='Countries', loc='upper right')
plt.title('Countries with Highest Responses')
plt.show()

# 4
df.iloc[:, 10:] = df.iloc[:, 10:].replace({'tick': True, 0: False})

df_september = df[df['Country'] != 'Egypt']

numeric_columns = df_september.iloc[:, 10:].apply(pd.to_numeric, errors='coerce')

respo = numeric_columns.sum()

respo.plot(kind='bar', color='crimson', edgecolor='black')
plt.title('Courses with Most Responses outside Egypt')
plt.xlabel('Courses')
plt.ylabel('No. of responses')
plt.xticks(rotation=50, ha='right')
plt.show()

# 5
important_resp = df['Is online learning crucial?'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(important_resp, labels=None, startangle=90,  autopct='%0.0f%%')
plt.legend(important_resp.index, title='Is online learning crucial?', loc='upper right')
plt.title('% of people thinking that online learning is important?')
plt.show()
```
